# House-Price-Prediction

## objective

Predicting the Price of a house using linear Regression

## Step 1: Importing Libraries

```python
# Data Handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt

# Dataset
from sklearn.datasets import fetch_california_housing

# Train/Test Split
from sklearn.model_selection import train_test_split

# Machine Learning Model
from sklearn.linear_model import LinearRegression

# Feature Scaling (We'll use this later in the project)
from sklearn.preprocessing import StandardScaler

# Evaluation Metrics
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import matplotlib.pyplot as plt
```

## Step 2: Fetching The Data
using california housing of sklearn as the DataSet

```python
housing = fetch_california_housing(as_frame=True)
df = housing.frame
```

## Step 3: Visualizing the first 5 rows

```python
df.head()
```

## Step 4: Checking Null Values

```python
df.isnull().sum()
```

## Step 5: Checking The Correlation of Features

```python
df.corr(numeric_only=True)
```

## Step 6: Visualizing Graph Between MedInc and MedHouseVal

```python
plt.scatter(df["MedInc"], df["MedHouseVal"])
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.show()
```

## Step 7: Model Training

```python
X = df.drop('MedHouseVal',axis = 1)
y = df['MedHouseVal']

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.8,random_state = 42)

model = LinearRegression()

model.fit(X_train,y_train)

ypred = model.predict(X_test)


```

## Step 8: Results

```python
mae = mean_absolute_error(y_test,ypred)
mse = mean_squared_error(y_test,ypred)
r2 = r2_score(y_test,ypred)


print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R2 score: {r2}")
```

## Step 9: Standardizing the Features

```python
#standardiziing
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Step 10: Traning the Model Again

```python
#Testing Again with standard scaler
model.fit(X_train,y_train)
ypred = model.predict(X_test)

mae = mean_absolute_error(y_test,ypred)
mse = mean_squared_error(y_test,ypred)
r2 = r2_score(y_test,ypred)


print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R2 score: {r2}")

```

## Step 11: Results with Standard Scaler

```python
mae = mean_absolute_error(y_test,ypred)
mse = mean_squared_error(y_test,ypred)
r2 = r2_score(y_test,ypred)


print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R2 score: {r2}")
```
